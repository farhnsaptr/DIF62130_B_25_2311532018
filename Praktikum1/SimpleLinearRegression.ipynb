````python
import pandas as pd

data = {'luas' : [6,8,10,14,16],
        'harga' : [7,9,13,17.5,18]}
df = pd.DataFrame(data)
print(df)
````

# Penjelasan Kode
import pandas as pd = Mengimpor library pandas dan memberikan alias pd. pandas adalah library Python yang sangat populer untuk manipulasi dan analisis data.

data = {'luas' : [6,8,10,14,16], 'harga' : [7,9,13,17.5,18]} = Membuat sebuah dictionary dengan dua kunci, yaitu 'luas' dan 'harga'. Setiap kunci memiliki nilai berupa list yang berisi data numerik. Dictionary ini akan digunakan sebagai data untuk membuat DataFrame.

df = pd.DataFrame(data) = Membuat sebuah DataFrame dari dictionary data. DataFrame adalah struktur data dua dimensi yang disediakan oleh pandas, mirip dengan tabel dalam database atau spreadsheet.

print(df) = Memberikan output dari df

````python
import matplotlib.pyplot as plt

df.plot(kind='scatter', x='luas', y='harga')
plt.title('Perbandingan Luas Tanah dengan Harga Tanah')
plt.xlabel('Luas Tanah (m^2)')
plt.ylabel('Harga Tanah ($)')
plt.xlim(0,25)
plt.ylim(0,25)
plt.grid(True)
plt.show()
````

````python
import numpy as np

x = np.array(df['luas'])
y = np.array(df['harga'])

print(f'X: {x}')
print(f'Y: {y}')
````

````python
x = x.reshape(-1, 1)

print(f'X Shape: {x.shape}')
print(x)
````

````python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x, y)
````

````python
X_vis = np.array([0, 25]).reshape(-1, 1)
y_vis = model.predict(X_vis)
````

````python
plt.scatter(x, y)
plt.plot(X_vis, y_vis, color='red')
plt.title('Perbandingan Luas Tanah dengan Harga Tanah')
plt.xlabel('Luas Tanah (m^2)')
plt.ylabel('Harga Tanah')
plt.grid(True)
plt.show()
````

````python
print(f'intercept: {model.intercept_}')
print(f'slope: {model.coef_}')
````

````python
X_train = np.array([6, 8, 10, 14, 18]).reshape(-1,1)
y_train = np.array([7, 9, 13, 17.5, 18])

X_test = np.array([9, 9, 11, 16, 12]).reshape(-1,1)
y_test = np.array([11, 8.5, 15, 18, 11])
````

````python
model = LinearRegression()
model.fit(X_train, y_train)
````

````python
from sklearn.metrics import r2_score

y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')
````
